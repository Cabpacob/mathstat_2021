```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
gamma = 0.7

def a_len(N):
    N = int(N)
    xs = np.random.normal(size=N)
    
    t_1 = np.sum(xs ** 2) / scipy.stats.chi2.ppf((1 + gamma) / 2, N)
    t_2 = np.sum(xs ** 2) / scipy.stats.chi2.ppf((1 - gamma) / 2, N)
    
    return t_2 - t_1

def b_len(N):
    N = int(N)
    xs = np.random.normal(size=N)
    
    t_1 = N * xs.mean() ** 2 / (scipy.stats.norm.ppf((3 + gamma) / 4)) ** 2
    t_2 = N * xs.mean() ** 2 / (scipy.stats.norm.ppf((3 - gamma) / 4)) ** 2
    
    return t_2 - t_1
```

```python
N = 1000

xs = np.linspace(1, N, 100)
ys = list(map(a_len, xs))

fig, ax = plt.subplots()

plt.plot(xs, ys, label='a')

ys = list(map(b_len, xs))
plt.plot(xs, ys, label='b')

plt.yscale('log')

plt.grid()
plt.legend()
plt.show()
```
